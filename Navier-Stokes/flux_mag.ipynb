```python
import numpy as np
import matplotlib.pyplot as plt


#densitat de flux en funció de l'alçada

h= 0.015 #alçada de l'imant (m)
r= 0.007 #radi de l'imant (m)
Br = 1.45 #flux magnetic romanent (T)
y = np.linspace(0, 0.03, 100) #alçada de la cubeta
B = (Br/2)*(((y+h)/np.sqrt((y+h)**2+r**2))-(y/np.sqrt(y**2+r**2)))


plt.figure(figsize=(7,7))
plt.style.use('ggplot')
plt.plot(y, B,color='teal')
plt.yticks(np.arange(0,0.7,0.05))
plt.xlabel('y (m)', fontsize=17)
plt.ylabel('Densitat de flux magnètic (T)',fontsize=17)
plt.show()


```

```python
#gradient de la densitat de flux magnètic
h= 0.015 #alçada de l'imant (m)
r= 0.007 #radi de l'imant (m)
Br = 1.45 #flux magnetic romanent (T)
y = np.linspace(0, 0.03, 100) #alçada de la cubeta
nablaB = ((Br*r**2)/2)*((pow((y+h)**2+r**2,-3/2))-(pow(y**2+r**2,-3/2)))

plt.figure(figsize=(7,7))
plt.plot(y, nablaB,color='teal', label= 'Solució numèrica')
plt.style.use('ggplot')
plt.xlabel('y (m)', fontsize=17)
plt.ylabel('Gradient de densitat de flux magnètic (T/m)',fontsize=17)
plt.yticks(np.arange(-100,10,10))
plt.show()

y_0=0
delta_B=((Br*r**2)/2)*((pow((y_0+h)**2+r**2,-3/2))-(pow(y_0**2+r**2,-3/2)))
print(delta_B)
```
